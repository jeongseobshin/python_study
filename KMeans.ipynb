KMeans datamining

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(columns=('x','y'))
```

```python
df.loc[0] = [2,3]
df.loc[1] = [2,6]
df.loc[2] = [2,8]
df.loc[3] = [4,5]
df.loc[4] = [4,6]
df.loc[5] = [4,8]
df.loc[6] = [5,2]
df.loc[7] = [5,5]
df.loc[8] = [5,1]
df.loc[9] = [5,6]
df.loc[10] = [6,8]
df.loc[11] = [6,9]
df.loc[12] = [6,4]
df.loc[13] = [7,5]
df.loc[14] = [7,2]
df.loc[15] = [7,3]
```

```python
df.head(16)
```

```python
sb.lmplot('x','y', data=df, fit_reg=False, scatter_kws={"s":100})
plt.title('K-means Example')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
points = df.values
kmeans = KMeans(n_clusters=4).fit(points)
kmeans.cluster_centers_
```

```python
kmeans.labels_ #데이터 분류
```

```python
df['cluster'] = kmeans.labels_
df.head(16)
```

```python
sb.lmplot('x','y', data=df, fit_reg=False, scatter_kws={"s":150}, hue="cluster") #hue->색넣기
plt.title('K-means Example')
plt.xlabel('X')
plt.ylabel('Y')
```
